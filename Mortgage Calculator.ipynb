# Mortage Calculator

```python
import tkinter as tk
from tkinter import messagebox
import math

def calculate_monthly_payment():
    try:
        principal = float(entry_principal.get())
        annual_interest_rate = float(entry_interest_rate.get())
        years = int(entry_years.get())
        
        monthly_interest_rate = annual_interest_rate / 12 / 100
        number_of_payments = years * 12
        
        if annual_interest_rate == 0:
            monthly_payment = principal / number_of_payments
        else:
            monthly_payment = principal * (monthly_interest_rate * math.pow(1 + monthly_interest_rate, number_of_payments)) / (math.pow(1 + monthly_interest_rate, number_of_payments) - 1)
        
        messagebox.showinfo("Monthly Payment", f"The monthly payment is: ${monthly_payment:.2f}")
    
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numbers for all fields.")

def clear_fields():
    entry_principal.delete(0, tk.END)
    entry_interest_rate.delete(0, tk.END)
    entry_years.delete(0, tk.END)

# Create the main window
root = tk.Tk()
root.title("Mortgage Calculator")

# Create and place the input fields and labels
tk.Label(root, text="Loan Amount (Principal):").grid(row=0, column=0, padx=10, pady=5)
entry_principal = tk.Entry(root)
entry_principal.grid(row=0, column=1, padx=10, pady=5)

tk.Label(root, text="Annual Interest Rate (%):").grid(row=1, column=0, padx=10, pady=5)
entry_interest_rate = tk.Entry(root)
entry_interest_rate.grid(row=1, column=1, padx=10, pady=5)

tk.Label(root, text="Loan Term (Years):").grid(row=2, column=0, padx=10, pady=5)
entry_years = tk.Entry(root)
entry_years.grid(row=2, column=1, padx=10, pady=5)

# Create and place the buttons
calculate_button = tk.Button(root, text="Calculate Monthly Payment", command=calculate_monthly_payment)
calculate_button.grid(row=3, column=0, pady=10)

clear_button = tk.Button(root, text="Clear", command=clear_fields)
clear_button.grid(row=3, column=1, pady=10)

# Run the application
root.mainloop()
```
